```python
import matplotlib.pyplot as plt

alpha = .01
o_bar = 0
o_bars = []
betas = []
for i in range(1000):
    o_bar = o_bar + alpha * (1 - o_bar)
    beta = alpha / o_bar
    o_bars.append(o_bar)
    betas.append(beta)

plt.figure(figsize=(10, 6))
plt.plot(o_bars, 'g-', linewidth=2, label='o_bar')
plt.plot(betas, 'r-', linewidth=2, label='beta')
plt.legend()
plt.xlabel('Steps')
plt.ylabel('Value')
```
